#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook

sinasc_RO = pd.read_csv('b1abb33c4f.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(5, 1)

sns.countplot(ax = axes[0], x='APGAR1', data=sinasc_RO)
sns.countplot(ax = axes[1], x='APGAR5', data=sinasc_RO)
sns.countplot(ax = axes[2], x='SEXO', data=sinasc_RO)
sns.countplot(ax = axes[3], x='GESTACAO', data=sinasc_RO)
sns.countplot(ax = axes[4], x='ESTCIVMAE', data=sinasc_RO)

```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(1, 2)

sns.countplot(ax = axes[0], x='CONSULTAS',data=sinasc_RO)
sns.barplot(ax = axes[1], x='CONSULTAS', y='APGAR5', data=sinasc_RO)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(12, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc_RO)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc_RO)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc_RO)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc_RO)
```
